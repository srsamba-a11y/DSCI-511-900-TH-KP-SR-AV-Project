## Abstract (General Audience)

Healthcare costs directly affect household finances, shaping access to care, financial security, and overall well-being. While much public discussion focuses on government healthcare spending, individuals in every healthcare system still bear costs through insurance premiums, copayments, deductibles, and direct payments for services. This project proposes the construction of a cross-national dataset comparing consumer healthcare expenditures across the United States, Canada, the United Kingdom, Japan, and China. By integrating data from international organizations and national statistical agencies, we aim to create a standardized dataset that captures how much people pay for healthcare out of their own pockets, enabling meaningful computational comparisons across countries with very different healthcare systems.

## 2. Project Topic and Motivation
### 2.1 Why We Chose This Topic

We’re curious about how much people actually spend on healthcare, not what the government spends. Everyone’s experience is different depending on where they live, and some systems make people pay a lot more out of pocket than others.

### 2.2 What We Hope to Learn

By comparing consumer healthcare costs across these five countries, we hope to see:

* How much people pay out-of-pocket for medical care
* Differences between universal healthcare and mixed systems
* Whether higher personal costs are related to system structure, not just wealth

---

## 3. Dataset Overview

### 3.1 What It Will Include

We plan to collect data about:
* Out-of-pocket healthcare spending per person
* Household healthcare spending as a percentage of total expenses
* Insurance premiums (if available)
* Copayments and deductibles
* Prescription drug spending
* Some basic info like population or income per capita
  
Each row will be country + year, so we can compare across countries and see trends over time.

### 3.2 Where We'll Get the Data

* WHO Global Health Expenditure Database
* OECD Health Statistics
* World Bank Open Data
* National statistics offices:
    * U.S. (Consumer Expenditure Survey)
    * Canada (Statistics Canada)
    * U.K. (ONS)
    * Japan (Ministry of Health)
    * China (National Bureau of Statistics)

Some of this data (like WHO and World Bank) is guaranteed to be accessible, so we know the project is doable.

### 5.2 Sample Data Code

**Sample consumer healthcare expenditure data**

import requests 
import pandas as pd 
import matplotlib.pyplot as plt

countries = {"United States": "USA", "Canada": "CAN", "United Kingdom": "GBR", "Japan": "JPN", "China": "CHN"}

indicator = "SH.XPD.OOPC.PP.CD"

base_url = ("https://api.worldbank.org/v2/country/{}/indicator/{}" "?format=json&per_page=5000")

all_data = []

for country_name, country_code in countries.items():
    url = base_url.format(country_code, indicator)
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Error fetching data for {country_name}")
        continue

    json_data = response.json()

    if len(json_data) < 2 or json_data[1] is None:
        print(f"No data available for {country_name}")
        continue

    for entry in json_data[1]:
        all_data.append({
            "country": country_name,
            "year": int(entry["date"]),
            "out_of_pocket_ppp": entry["value"]
        })
        
df = pd.DataFrame(all_data)

df = df.sort_values(by=["country", "year"])

df_2020_2024 = df[(df["year"] >= 2020) & (df["year"] <= 2024)].dropna(subset=["out_of_pocket_ppp"])

print("Filtered data (2020–2024):")
print(df_2020_2024)

plt.figure()

for country in df_2020_2024["country"].unique():
    subset = df_2020_2024[df_2020_2024["country"] == country]
    plt.plot(subset["year"], subset["out_of_pocket_ppp"], marker="o", label=country)

plt.title("Out-of-Pocket Health Expenditure per Capita (PPP), 2020–2024")
plt.xlabel("Year")
plt.ylabel("PPP (International $)")
plt.legend()
plt.grid(True)
plt.show()

df_2020_2024.to_csv("out_of_pocket_health_expenditure_ppp_2020_2024.csv", index=False)

print("CSV saved as: out_of_pocket_health_expenditure_ppp_2020_2024.csv")

## 7. Limitations

* Countries define “out-of-pocket” differently
* China’s data might be harder to get
* Some households may underreport spending
* Data might not be available for every year

We’ll be transparent about these issues in the README and data dictionary.

## 9. Limitations and Opportunities for Improvement

Expected limitations include:

* Differences in accounting standards across countries
* Incomplete time coverage for some countries

Future improvements could include adding more countries, incorporating health outcomes, or linking cost data with demographic indicators.
